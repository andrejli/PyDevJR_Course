```python
import matplotlib.pyplot as mp
import numpy as np
import pandas as pd
```

```python
mp.plot([1,2,3,4,5], [10,20,30,40,50], [1,2,3,4,5], [78, 45, 67, 34, 56], [1,2,3,4,5], [45, 34, 23, 11, 0])
mp.title('NAS GRAF')
mp.xlabel('Vzorka c.')
mp.ylabel('Hodnota v J')
mp.show()
```

```python
mp.figure(figsize=(15,3))
mp.plot([1,2,3,4,5], [10,20,30,40,50], 'bo')
mp.plot([1,2,3,4,5], [23, 45,12,55,3], 'red')
mp.title('NAS GRAF')
mp.xlabel('Vzorka c.')
mp.ylabel('Hodnota v J')
mp.show()
```

## BAR CHART

```python
mp.figure(figsize=(8,3))
mp.bar([1,2,3,4,5], [23,67,90,15,19])
mp.show()
```

```python
divisions = ['DivA', 'DivB', 'DivC', 'DivD', 'DivE']
values = [10,20,30,40,50]
namerane_hodnoty = [8, 28,33,41,59]
index = np.arange(5)
width = .3
mp.figure(figsize=(8,5))
mp.bar(index, values, width, color='green', label='Predpokladane hodnoty')  # graf 1
mp.bar(index, namerane_hodnoty, width, color='blue', label='Namerane hodnoty', bottom=values)  # graf 2
mp.xticks(index, divisions)
mp.legend(loc='best')
mp.show()
```

# KOLACOVE GRAFY


```python
mp.figure(figsize=(15,5))
firmy = ['A', 'B', 'C', 'D', 'E']
podiel_na_trhu = [45,35,10,5,5]
exploded = [0,0,1,0,0]
mp.pie(podiel_na_trhu, explode=exploded, labels=firmy, shadow=True, startangle=45)
mp.legend(title='Podiel na trhu', loc='upper right')
mp.title('NAS KOLAC')
mp.show()
```

```python
mp.figure(figsize=(15,5))
mp.subplot(1,3,1)
mp.pie([10,20,30,40,50], shadow=True)
mp.title('PRVY GRAF')

mp.subplot(1,3,2)
mp.barh([1,2,3,4], [10,20,30,40], color='orange')
mp.title('DRUHY GRAF')

mp.subplot(1,3,3)
mp.bar([1,2,3,4], [10,20,30,40], color='orange')
mp.title('DRUHY GRAF')

mp.suptitle('DVOJICA GRAFOV')
mp.show()
```

# HISTOGRAM

```python
mp.figure(figsize=(10,5))
x = np.random.randn(10000)
mp.hist(x,500)
mp.title('NAS HISTOGRAM')
mp.show()
```

## SCATTER GRAPH



```python
a = pd.Series([1,2,3,4,5])
b = pd.Series([1,12,3,24,15])
mp.xlim(0,5)
mp.ylim(0,30)
mp.scatter(a,b)
mp.show()
```
